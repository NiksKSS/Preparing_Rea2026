```python
df = pd.read_csv('Air_Quality.csv')
```

```python
X = df.drop(['Date', 'City', 'AQI'], axis=1)
y = df['AQI']
```

```python
X_train, X_val , y_train, y_val = train_test_split(X, y, test_size=0.15)

X_test, X_ft , y_test, y_ft = train_test_split(X_val, y_val, test_size=0.25)
```

```python
scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
X_ft = pd.DataFrame(scaler.transform(X_ft), columns=X_ft.columns)
```

```python
X_train
```

```python
joblib.dump(X_train, 'old_X_train_reg.pkl')
joblib.dump(y_train, 'old_y_train_reg.pkl')
joblib.dump(X_test, 'X_test_reg.pkl')
joblib.dump(y_test, 'y_test_reg.pkl')
```
