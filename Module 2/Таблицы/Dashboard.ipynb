# Формирование интерактивной панели (dashboard)

Часто на чемпионатах могут дать такое задание на 2 модуле:

###

Для своевременного принятия решений в области экологического мониторинга, минимизации рисков для здоровья населения и оперативного реагирования на ухудшение качества воздуха необходимо построить аналитический дашборд загрязнённости воздуха.

Интерактивная информационная панель должна наглядно представлять, визуализировать и анализировать данные о качестве атмосферного воздуха. Для этого необходимо подключить к системе источник данных (датасет загрязнённости воздуха или базу данных), настроить регламент обновления данных и уровни доступа пользователей.

Аналитическая система должна позволять пользователю:

- автоматизировать работу с экологическими данными;

- анализировать показатели загрязнённости в реальном времени или вблизи реального времени;

- видеть ключевые экологические метрики на одном экране;

- легко считывать информацию, выявлять тенденции и делать выводы о состоянии воздуха.



Разработанный дашборд должен позволять проводить анализ качества и загрязнённости воздуха в различных локациях в течение выбранного временного интервала.

Необходимо вычислить и отобразить следующие характеристики:

1. уровень загрязнения воздуха в процентах относительно предельно допустимой концентрации (ПДК) для выбранных загрязняющих веществ;

2. фактические концентрации основных загрязнителей (например, PM2.5, PM10, NO₂, SO₂, CO, O₃) за заданный временной интервал;

3. топ наиболее загрязнённых станций (или районов) за выбранный период;

4. средние значения концентраций загрязняющих веществ по станциям за заданный интервал времени;

5. количество станций, на которых превышены нормативы ПДК хотя бы по одному показателю;

###

Построение дашборда происходит в файле dashboard.py

Под дашборд берем связку фреймворка streamlit и библиотеки plotly.

Запуск с помощью команды:

```
streamlit run dashboard.py
```
